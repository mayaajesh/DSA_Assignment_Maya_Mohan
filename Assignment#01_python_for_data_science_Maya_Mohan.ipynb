# Data Visualization

```python
sns.set_theme(style="darkgrid")
sns.countplot(x='Classification',data=data_f)
plt.title('Count of classification')
```

```python
sns.boxplot(x=data_f['Classification'], y=data_f["SL"])
plt.title('Sepal Length of 3 different classifications')
```

```python
sns.boxplot(x=data_f['Classification'], y=data_f["SW"], palette="Blues")
```

```python
sns.scatterplot(x=data_f['SL'],y=data_f['SW'],hue=data_f['Classification'])
plt.title('Scatter plot based on sepal width and sepal length')
```

```python
sns.scatterplot(x=data_f['PL'],y=data_f['PW'],hue=data_f['Classification'])
plt.title('Scatter plot based on petal width and petal length')
```
